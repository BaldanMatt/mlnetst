# Case study: Interactions between cells in space
We draw Graphs of the cells in a slice of tissue

```python
# Read the graph
G_full = nx.read_graphml(media_dir / "mouse1_slice153_spatial_graph.graphml")
# Subsample nodes
N = 1000
if N:
    sub_nodes = random.sample(list(G_full.nodes()), N)
    G = G_full.subgraph(sub_nodes).copy()
else:
    G = G_full
# Position and color extraction
pos = {n: (float(attrs["x"]), float(attrs["y"])) for n, attrs in G.nodes(data=True)}
node_labels = [attrs.get("label", "Other") for _, attrs in G.nodes(data=True)]

# Prepare node DataFrame
nodes = list(G.nodes())
node_coords = np.array([pos[n] for n in nodes])
node_df = pd.DataFrame({
    'x': node_coords[:, 0],
    'y': node_coords[:, 1],
    'index': nodes,
    'label': node_labels
})

# Prepare edge path data
edge_paths = []
for u, v in G.edges():
    try:
        x0, y0 = pos[u]
        x1, y1 = pos[v]
        edge_paths.append([(x0, y0), (x1, y1)])
    except KeyError:
        continue  # In case a node is missing from pos dict

# Build HoloViews elements
hv_edges = hv.Path(edge_paths)
hv_nodes = hv.Points(node_df, kdims=["x", "y"], vdims=["index", "label"])

# Efficient rasterization using datashader
# Apply datashader without unused params
shaded_edges = datashade(hv_edges, cmap=["gray"])
color_dict={
    "Glutamatergic": hex_to_rgb("#0000FF"),
    "GABAergic": hex_to_rgb("#FF0000"),
    "Other": hex_to_rgb("#00FF00")
}
shaded_nodes = dynspread(datashade(hv_nodes, aggregator=count_cat("label"), color_key=color_dict), threshold=0.5, max_px=3)

# Combine
graph = shaded_edges * shaded_nodes
graph.opts(width=1200, height=1000, xaxis=None, yaxis=None, tools=["hover"])
```
